# model design
 * Y^=x*w +b
 * random W for initial
 * loss = (Y^ - y)^2   (Mean Square Error)

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

def forward(x):
  return x*w

def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) * (y_pred - y)

w_list = []
mse_list = []
for w in np.arange(0.0, 4.0, 0.1):
  print('w=', w)
  l_sum = 0
  for x_val, y_val in zip(x_data, y_data):
    y_pred_val = forward(x_val)
    loss_val = loss(x_val, y_val)
    l_sum += loss_val
    print('\t', x_val, y_val, y_pred_val, loss_val)
  print('MSE=', l_sum/3)
  w_list.append(w)
  mse_list.append(l_sum/3)


````

````python
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
````

````python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0

def forward(x):
  return x * w

def cost(xs, ys):
  cost = 0
  for x, y in zip(xs, ys):
    y_pred = forward(x)
    cost += (y_pred - y)**2
  return cost / len(xs)

def gradient(xs, ys):
  grad = 0
  for x, y in zip(xs, ys):
    grad += 2* x * (x * w -y)
  return grad / len(xs)

print('Predict (before training)', 4, forward(4))

epochs = []
losses = []

for epoch in range(100):
  cost_val = cost(x_data, y_data)
  grad_val = gradient(x_data, y_data)
  w -= 0.01* grad_val

  epochs.append(epoch)
  losses.append(cost_val)
  print('epoch:', epoch, 'w=', w, 'loss=', cost_val)
print('Predict (after training)', 4, forward(4))


````

````python
plt.plot(epochs, losses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.title('Epoch vs Loss')
plt.show()
````

# Stochastic Gradient Descent

````python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0

def forward(x):
    return x * w

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2

def gradient(x, y):
    return 2* x * (x * w -y)

print('Predict (before training)', 4, forward(4))

epochs = []
losses = []

for epoch in range(100):
  for x, y in zip(x_data, y_data):
    grad = gradient(x, y)
    w = w - 0.01 * grad

    print('\tgrad:', x, y, grad)

    l= loss(x, y)
  print('progress:', epoch, 'w=', w, 'loss=', l)

  epochs.append(epoch)
  losses.append(l)
print('Predict (after training)', 4, forward(4))


````

````python
plt.plot(epochs, losses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.title('Epoch vs Loss')
plt.show()
````

# Compute gradient in simple network

# use chain rule to compute the gradient (backward)

````python
import torch

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = torch.Tensor([1.0])
w.requires_grad = True

def forward(x):
    return x * w

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2

print('Predict (before training)', 4, forward(4).item())

epochs = []
losses = []

for epoch in range(100):
  for x, y in zip(x_data, y_data):
    l= loss(x, y)
    l.backward()

    print('\tgrad:', x, y, w.grad.item())
    w.data = w.data - 0.01 * w.grad.data

    w.grad.data.zero_()

    epochs.append(epoch)
    losses.append(l.data)
  print('progress:', epoch, l.item())


print('Predict (after training)', 4, forward(4).item())
````

````python
plt.plot(epochs, losses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.title('Epoch vs Loss')
plt.show()
````

````python
import torch

x_data = torch.Tensor([[1.0], [2.0], [3.0]])
y_data = torch.Tensor([[2.0], [4.0], [6.0]])


class LinearModel(torch.nn.Module):
  def __init__(self):
      super(LinearModel, self).__init__()
      self.linear = torch.nn.Linear(1, 1) # weight and bias(1, 1)

  def forward(self, x):
    y_pred = self.linear(x)
    return y_pred

model = LinearModel()

criterion = torch.nn.MSELoss(size_average=False)

optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

epochs = []
losses = []

for epoch in range(1000):
  y_pred = model(x_data)
  loss = criterion(y_pred, y_data)
  print (epoch, loss.item())

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()


  epochs.append(epoch)
  losses.append(loss.item())

print('w=', model.linear.weight.item())
print('b=', model.linear.bias.item())

x_test = torch.Tensor([[4.0]])
y_test = model(x_test)
print('y_pred = ', y_test.data)
````

````python
import torch.nn.functional as F
import torch

x_data = torch.Tensor([[1.0], [2.0], [3.0]])
y_data = torch.Tensor([[0], [0], [1]])

class LogisticRegressionModel(torch.nn.Module):
    def __init__(self):
        super(LogisticRegressionModel, self).__init__()
        self.linear = torch.nn.Linear(1, 1)

    def forward(self, x):
        y_pred = F.sigmoid(self.linear(x))
        return y_pred


model = LogisticRegressionModel()

criterion = torch.nn.BCELoss(size_average=False)

optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

epochs = []
losses = []

for epoch in range(1000):
  y_pred = model(x_data)
  loss = criterion(y_pred, y_data)
  print (epoch, loss.item())

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  epochs.append(epoch)
  losses.append(loss.item())

x_test = torch.Tensor([[4.0]])
y_test = model(x_test)
print('y_pred = ', y_test.data)

````

````python
import torch.nn.functional as F
import torch
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import random_split, TensorDataset, Dataset, DataLoader

class DiabetesDataset(Dataset):
  def __init__(self):
     pass

  def __getitem__(self, index):
     pass

  def __len__(self):
    pass

dataset = DiabetesDataset()
train_loader = DataLoader(dataset=dataset, batch_size=32, shuffle=True, num_workers=2)

# 
xy = np.loadtxt('diabetes.csv.gz', delimiter=',', dtype=np.float32)
x_data = torch.from_numpy(xy[:, :-1])  # 
y_data = torch.from_numpy(xy[:, [-1]])  # 

# Create a TensorDataset
dataset = TensorDataset(x_data, y_data)

# Split into training and validation sets (80% train, 20% val)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# Extract data from the subsets
train_x = torch.stack([item[0] for item in train_dataset])  # Extract features
train_y = torch.stack([item[1] for item in train_dataset])  # Extract labels
val_x = torch.stack([item[0] for item in val_dataset])
val_y = torch.stack([item[1] for item in val_dataset])

# 
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear1 = torch.nn.Linear(8, 6)
        self.linear2 = torch.nn.Linear(6, 4)
        self.linear3 = torch.nn.Linear(4, 1)
        self.sigmoid = torch.nn.Sigmoid()

    def forward(self, x):
        x = self.sigmoid(self.linear1(x))
        x = self.sigmoid(self.linear2(x))
        x = self.sigmoid(self.linear3(x))
        return x

model = Model()

# 
criterion = torch.nn.BCELoss()  # Binary Cross Entropy Loss
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

# 
epochs = []
train_losses = []
val_losses = []

for epoch in range(1000):
    # ------------------
    # 
    # ------------------
    model.train()
    y_pred = model(train_x)  # 
    train_loss = criterion(y_pred, train_y)  # 

    optimizer.zero_grad()  # 
    train_loss.backward()  # 
    optimizer.step()  # 

    # ------------------
    # 
    # ------------------
    model.eval()  # 
    with torch.no_grad():  # 
        val_pred = model(val_x)
        val_loss = criterion(val_pred, val_y)

    # 
    print(f"Epoch {epoch + 1}: Train Loss: {train_loss.item():.4f}, Val Loss: {val_loss.item():.4f}")

    # 
    epochs.append(epoch)
    train_losses.append(train_loss.item())
    val_losses.append(val_loss.item())

# 
plt.plot(epochs, train_losses, label="Train Loss")
plt.plot(epochs, val_losses, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Train vs Validation Loss")
plt.legend()
plt.grid()
plt.show()

# 
x_test = torch.tensor([[0.5, 0.2, 0.1, 0.3, 0.4, 0.5, 0.6, 0.7]])  # 
y_test = model(x_test)
print('Test Input:', x_test)
print('Predicted Output:', y_test.data)
````
